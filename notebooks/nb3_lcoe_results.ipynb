```python
from ass1.modeling.loaders import load_tmy
from ass1.modeling.location import WindResourceData

df = load_tmy()
loc = WindResourceData(df["datetime"], df["WS50M"], df["WD50M"])
resource = loc.create_weibull_wind_resource()
```

```python
from ass1.modeling.financial import FinancialModel
import pandas as pd

from ass1.modeling.loaders import load_nrel_6MW, load_nrel_8MW, load_nrel_10MW, load_dtu_10MW

fcr = 0.05

turbine_options = [
    load_nrel_6MW(), load_nrel_8MW(),load_nrel_10MW(),
    load_dtu_10MW()
    ]

finance = FinancialModel(fcr=fcr)

results = {}
for turbine in turbine_options:
    aep = resource.aep(turbine)
    c_rotor, c_tower, c_other = finance.cost_turbine_components(turbine)
    lcoe = finance.lcoe(turbine, aep)

    results[turbine.name] = {
        "Anunual Energy Production (GWh)": aep * 1e-6,
        "Rotor + Tower cost (AUD)": c_rotor+c_tower,
        "Other costs (AUD)": c_other,
        "TCC (AUD)": c_rotor + c_tower +c_other,
        "Levelised Cost Energy (AUD/MWh)": lcoe * 1e3
    }

df = pd.DataFrame(results).T
df
```
